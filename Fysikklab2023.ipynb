## Teori
I simuleringen av forsøket gjøres en rekke antakelser om det fysiske systemet. Den grunnleggende antakelsen er at kula ruller rent. Det vil si at det ikke er noen form for friksjonstap mellom kula og underlaget i rullebevegelsen og legger grunn for teorien i dette forsøket. 

Dette medfører at rullebetingelsen
$$\omega = \frac{V}{R} \label{eq1}\tag{1},$$

er oppfylt. Her er $\omega$ vinkelfart, $V$ er translasjonsfarten til massesenteret og $R$ er radien i kula. Under antakelsen om at kula ruller rent og ikke opplever luftmotstand, vil det ikke være tap av energi til ytre krefter og den totale mekaniske energien er bevart, dvs. at kulas startenergi og sluttenergi er lik. Dette gjør det mulig å finne fart som funksjon av $x$ ved energibevaring. Bevaring av mekanisk energi gir
$$mgy_0+\frac{1}{2}mv_0^2+\frac{1}{2}cmv_0=mgy(x)+\frac{1}{2}mv(x)^2+\frac{1}{2}cmv(x)^2 \label{eq2}\tag{2},$$

der m er kulas masse, $g$ er gravitasjonsakselerasjonen, $y_0$ og $y(x)$ er henholdsvis starthøyde og momentan høyde, $v_0$ og $v(x)$ er henholdsvis startfart og momentan fart. Ved start $t=0$ har kula høyde $y(0) = y_0$ og fart $v(0)=v_0=0 m/s$ Total mekanisk energi ved start vil dermed være gitt av 
$$mgy_0 = mgy(x) +\frac{1}{2}mv(x)^2+\frac{1}{2}cmv(x)^2\label{eq3}\tag{3}.$$

Ved å løse for farten ved $c = 2/5$ for en kompakt kule får man sammenhengen under
$$v(x)=\sqrt{\frac{10g(y_0-y(x))}{7}}\label{eq4}\tag{4},$$
der $x$ er horisontal posisjon, $y_0$ er vertikal startposisjon og $y(x)$ er vertikal posisjon i punktet $(x, y(x))$ og $g$ er gravitasjonsakselerasjonen. 

Den maksimale friksjonen som til enhver tid kan virke på en kule som ruller rent i kontaktpunkt med banen er gitt av
$$f = \mu_sN\label{eq5}\tag{5},$$
der $\mu_s$ er den statiske friksjonskoeffisienten og $N$ er normalkraften. Her brukes den statiske friksjonskoeffisienten fordi ballens kontaktpunkt med banen alltid har fart lik 0 når rullebetingelsen er oppfylt. Positiv retning regnes for å være oppover og ved Newtons 2. lov normalt på banen finnes
$$\sum{F}=ma_\bot=N-mg \cos{\beta}\label{eq6}\tag{6},$$
hvor $a_\bot$ er sentripetalakselerasjonen i punktet på banen, $\beta(x)$ er banens helning i punktet, og det andre leddet i likningas høyre side er gravitasjonskrafts-komponenten normalt på banen.
Helningsvinkelen til banen finner man ved å se på den deriverte til banen,
$$y’(x)=\frac{dy}{dx}(x)=\tan{\beta}\label{eq7}\tag{7}.$$
Dette gir videre en enkel formel for $\beta$ som funksjon av $x$ ved
$$\beta(x)=\arctan{\frac{dy}{dx}(x)}\label{eq8}\tag{8}.$$
Her vil helningsvinkelen ha positivt fortegn for oppoverbakker og motsatt for nedoverbakker.
Sentripetalakselerasjonen er gitt ved
$$a_\bot=\frac{v(x)^2}{R}=v(x)^2\kappa\label{eq9}\tag{9},$$
der $\kappa$ er den inverse krumningsradien til banen i punktet $x$. $\kappa$ er gitt av 
$$\kappa=\frac{y’’(x)}{(1+y’(x)^2)^\frac{3}{2}}\label{eq10}\tag{10}.$$
Normalkrafta er dermed gitt ved
$$N=m(g\cos{\beta}+a_\bot)\label{eq11}\tag{11}.$$
Dette stemmer godt da normalkrafta vil bli større dersom banen krummer oppover og mindre dersom banen krummer nedover da sentripetalakselerasjonen vil peke henholdsvis oppover og nedover som øker og reduserer $N$. Videre kan en se på de kreftene som virker tangentielt på banen. Da vil gravitasjonens tangentielle komponent gitt ved 
$$-mg\sin{\beta}\label{eq12}\tag{12},$$
og friksjonskraften $f$ virke med og mot hverandre avhengig av banens helning. Dersom $\beta < 0$ altså nedover bakke vil gravitasjonskraften virke i positiv retning (nedover banen) og friksjonskrafta virker i negativ retning (mot bevegelsesretningen). $N2$ for de banetangentielle kreftene blir dermed
$$-mg\sin{\beta}+f=ma\label{eq13}\tag{13}.$$

Baneakselerasjonen finnes ved å derivere uttrykket for fart
$$a=\frac{dv}{dy}\frac{dy}{dt}=-\frac{g\sin{\beta}}{1+c}\label{eq14}\tag{14}.$$

Ved å sette denne inn i likninga for $N2$ tangentielt får $f$ uttrykket
$$f=ma+mg\sin{\beta}=\frac{c}{c+1}mg\sin{\beta}\label{eq15}\tag{15}.$$
Innsetting av verdien for $c=\frac{2}{5}$ blir $f=\frac{2mg\sin{\beta}}{7}.$

Videre vil det være interessant å vurdere hvorvidt antakelsen om ren rulling er tilstrekkelig. Ved å se på størrelsen $|\frac{f}{N}|$ over kulas forløp over banen kan en undersøke om denne antakelsen er tilstrekkelig god. Så lenge denne verdien ikke overstiger den statiske kule-bane-friksjonskoeffisienten vil antakelsen være god.

For å vurdere gyldigheten til de eksperimentelle dataene er det nødvendig å se til usikkerhet i målinger og dermed resultater. Uten et bevisst forhold til usikkerhet vil resultater kunne tas ut av kontekst og feiltolkes. Sentralt i usikkerhetsanalyse står middelverdi (forventningsverdi), standardavvik og standard feilen til målingene som utføres. Ved å ta flere målinger og se på middelverdien vil en få et bedre bilde av det forventede utfallet av forsøket. Middelverdien er gitt ved
$$\bar{x}=\frac{1}{N} \sum_{i=1}^{N} x_i \label{eq16}\tag{16}.$$
Dersom $N$ er stor nok, vil $\bar{x}$ kunne anses som forventningsverdien. Det er også interessant å se på hvordan en vilkårlig måling vil avvike fra den forventede verdien slik at en får et intervall hvor en kan anse målinger som troverdige. Dersom målinger får større avvik enn standardavviket vil det kunne være interessant å undersøke denne målingen med tanke på feil i forsøkets utførelse, etc. Standardavviket er gitt ved
$$\delta x=\sqrt{\frac{1}{N-1}\sum_{i=1}^{N}(x_i-\bar{x})^2}\label{eq17}\tag{17}.$$
Det kan også være interessant å se på feilen i gjennomsnitt. Dette kalles standardfeil og er gitt ved
$$\delta \bar{x} = \frac{\delta x}{\sqrt{N}}\label{eq18}\tag{18}.$$
Her ser en at dersom en tar flere målinger, $N$, vil feilen i gjennomsnittet avta og en får bedre kjennskap til den sanne verdien av $\bar{x}$ og dermed den forventede verdien av forsøksutfallet.

## Metode



Arbeidsmetoden i forsøket baserte seg på å regne ute teoretiske størrelser med ulike numeriske metoder i programmeringsspråket Python for deretter å utføre det reelle forsøket. For å kunne beregne teoretiske verdier for kula ved bruk av numeriske metoder, produserte vi en ideell modell av den fysiske baneformen, $y(x)$ ved bruk av et utlevert Pythonskript [3]. Pythonskriptet anvendte funksjonen CubicSpline fra SciPy-biblioteket for å interpolere mellom 8 skruefestepunkt ved bruk av tredjegradspolynomer, slik at banen blir kontinuerlig gjennom festepunktene: cs = CubicSpline(festepunkter_x,festepunkter_y,bc_type='natural'). festepunkter_x og festepunkter_y er tabeller med verdier (numpy arrays), mens bc_type='natural' gjør at krumningen er lik null i de to ytterste punkta. Selve skruefestepunkta ble generert tilfeldig for å produsere modellen for $y(x)$. Modellen gjorde at vi videre kunne estimere sentrale størrelser for banen og kula. Ved å bruke banens starthøyde, $y(0)$, kunne kulas fart i et vilkårlig punkt $(x,y(x))$ beregnes ut ifra $\ref{4}$, der $g=9.81m/s^2$. 

Med formlene for $v(x)$ og $y(x)$ kunne også total kinetisk og potensiell energi i start- og sluttpunktet beregnes. Ved å sammenlikne mekanisk energi i startpunktet og den mekaniske energien i sluttpunktet kunne vi sjekke om bevaring av mekanisk energi var oppfylt og dermed evaluere gyldigheten til antakelsene som ble gjort. 

Ved å ta i bruk kulas vertikale posisjon kunne banens helningsvinkel beregnes med formel *insert ref her*. Helningsvinkelen ble brukt til å beregne tidsutviklingen i simuleringen, blant annet kulas totale rulletid. For å beregne tidsutviklingen brukte vi en numerisk metode basert på energibevaring. Ettersom den simulerte banen er delt opp i 1401 jevnt fordelte x-verdier, vil avstanden mellom hver x tilsvare én millimeter da banen er 1400 mm lang. Med kjent hastighet $v_n$ og banevinkel $\beta_n$ i punktet $(x_n,y_n)$ der $n\in[1,2,…,1400]$ kunne vi beregne tiden $\Delta t_n$ kula brukte på et intervall mellom x_(n-1) og x_n. 

Horisontal hastighet i punkt x_n er gitt ved v_(x,n)=v_n cos⁡(b_n), så en god tilnærming av den gjennomsnittlige horisontalkomponenten til farten på intervall n blir 

⟨├ v_x ⟩_n= ┤  1/2(v_(x,n-1)+v_(x,n)).

Fra den grunnleggende definisjonen 〈v_x 〉_n= ∆x_n/∆t_n får vi dermed
∆t_n=  (∆x_n)/〈v_x 〉_n =  (2∆x_n)/(v_(x,n-1)+v_(x,n) ),
der hvert steg ∆x_n har lengde 1 mm. Kula, og dermed dens massesenter, starter i posisjon (x_0,y_0 ) med starthastighet v_0=0 i tidspunktet t_0=0. Dette betyr at den passerer punktene (x_n,y_n) med hastighet v_n i tidspunktene 
t_n= ∑_(j=1)^n▒〖∆t_j 〗 ; n=1,2,…,1400.

Ved å implementere pythonkode som numerisk regner ut lengden av alle tidsstegene kunne vi finne den totale rulletiden til kula i simuleringen. 

For å teste om antakelsen om ren rulling var gyldig, kunne vi sjekke forholdet mellom friksjonskraften og normalkraften. Dersom kravet om ren rulling skulle være oppfylt, ville

|(f(x))/(N(x))|≤μ_s, 
der f(x) er friksjonskraften mellom ballen og underlaget, og N(x) er normalkraften på kula. 
Her er μ_s friksjonskoeffisienten mellom kula og underlaget i en stillestående posisjon, med en antatt verdi på 0.4. Altså vil antakelsen om ren rulling være gyldig dersom forholdet mellom kreftene aldri overstiger 0.4 .  For å beregne N(x) og f(x), brukte vi *sett inn ref likninger*. Her ble helningsvinkelen brukt til å beregne y’(x) og y’’(x), slik at vi kunne finne et uttrykk for krumningen til grafen gitt ved *ref*. Krumningen ble videre brukt til å beregne sentripetalakselerasjon gitt ved *insert ref her*. Til slutt ble sentripetalakselerasjonen brukt til å finne verdier for normalkraften, N(x), og friksjonskraften, f(x).
